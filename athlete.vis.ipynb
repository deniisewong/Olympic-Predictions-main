```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
athletes = pd.read_csv('out/historical_athletes.csv')
athletes
```

```python
plt.figure(figsize=(10, 6))
plt.hist(athletes['weight'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Athlete Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(athletes['height'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Athlete Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(athletes['age'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Athlete Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
