```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('out/economic_data.csv')

data
```

```python
for series in data['Series'].unique():
    # Filter the DataFrame for the current Series
    series_df = data[data['Series'] == series]

    # Set the index to 'noc' and transpose the DataFrame to have years as rows
    series_df = series_df.set_index('noc').T

    # Drop the 'Series' row which is now a column due to the transpose
    series_df = series_df.drop('Series')

    # Convert the index to int for proper plotting
    series_df.index = series_df.index.astype(int)

    # Plot the data for each noc in the current Series
    plt.figure(figsize=(10, 6))
    for noc in series_df.columns:
        plt.plot(series_df.index, series_df[noc], label=noc)

    # Add title and labels
    plt.title(f'{series} over Time')
    plt.xlabel('Year')
    plt.ylabel(series)
    plt.legend(title='NOC')
    plt.show()
```
