```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

import pandas as pd
```

```python
data = pd.read_csv('out/historical_athletes.csv')
data
```

```python
data = pd.read_csv('out/historical_athletes.csv')
# convert medals, 1 = gold, 2 = silver, 3 = bronze
medal_mapping = {
    'Gold': 1,
    'Silver': 2,
    'Bronze': 3,
}
gender_mapping = {
    "Male": 0,
    "Female": 1
}

bool_mapping = {
    True: 1,
    False: 0
}

data['medal'] = data['medal'].map(medal_mapping)
data['medal'] = data['medal'].fillna(4)
data['gender'] = data['gender'].map(gender_mapping)
data['is_host'] = data['is_host'].map(bool_mapping)
data
```

```python
new_data = pd.read_csv('out/returning_athletes.csv')
new_data['gender'] = new_data['gender'].map(gender_mapping)
new_data['is_host'] = new_data['is_host'].map(bool_mapping)
new_X = new_data.drop(columns=['name', 'noc', 'Sport'])
```

```python
y = data['medal']
X = data.drop(columns=['medal', 'name', 'noc', 'sport']).values
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
rf_pipeline = make_pipeline(
    StandardScaler(),
    RandomForestClassifier(n_estimators=300, max_depth=30, min_samples_split=10, min_samples_leaf=4)
)
gb_pipeline = make_pipeline(
    StandardScaler(),
    GradientBoostingClassifier()
)
mlp_pipeline = make_pipeline(
    StandardScaler(),
    MLPClassifier(max_iter=500)
)
dt_pipeline = make_pipeline(
    StandardScaler(),
    DecisionTreeClassifier(max_depth=30, min_samples_split=10, min_samples_leaf=4)
)
svm_pipeline = make_pipeline(
    StandardScaler(),
    SVC()
)
knn_pipeline = make_pipeline(
    StandardScaler(),
    KNeighborsClassifier(n_neighbors=10)
)
```

```python
rf_pipeline.fit(X_train, y_train)
y_pred = rf_pipeline.predict(X_valid)
print("Classifier Report:")
print(classification_report(y_valid, y_pred))
print(f"Accuracy: {accuracy_score(y_valid, y_pred):.2f}")
```

```python
gb_pipeline.fit(X_train, y_train)
y_pred = gb_pipeline.predict(X_valid)
print("Classifier Report:")
print(classification_report(y_valid, y_pred))
print(f"Accuracy: {accuracy_score(y_valid, y_pred):.2f}")
```

```python
mlp_pipeline.fit(X_train, y_train)
y_pred = mlp_pipeline.predict(X_valid)
print("Classifier Report:")
print(classification_report(y_valid, y_pred))
print(f"Accuracy: {accuracy_score(y_valid, y_pred):.2f}")
```

```python
dt_pipeline.fit(X_train, y_train)
y_pred = dt_pipeline.predict(X_valid)
print("Classifier Report:")
print(classification_report(y_valid, y_pred))
print(f"Accuracy: {accuracy_score(y_valid, y_pred):.2f}")
```

```python
svm_pipeline.fit(X_train, y_train)
y_pred = svm_pipeline.predict(X_valid)
print("Classifier Report:")
print(classification_report(y_valid, y_pred, zero_division=0))
print(f"Accuracy: {accuracy_score(y_valid, y_pred):.2f}")
```

```python
knn_pipeline.fit(X_train, y_train)
y_pred = knn_pipeline.predict(X_valid)
print("Classifier Report:")
print(classification_report(y_valid, y_pred))
print(f"Accuracy: {accuracy_score(y_valid, y_pred):.2f}")
```

```python
# load
y_pred = knn_pipeline.predict(new_X.values)
y_pred

new_data['medal'] = y_pred
```
